# Exploratory Data Analysis(EDA) of Property Data

```python
# Import required libraries for EDA
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from warnings import filterwarnings
import warnings as wr
wr.filterwarnings('ignore')
```

```python
# Code to upload the data 
df=pd.read_csv('Property_data.csv')
```

```python
df.columns
```

## In the given property dataframe 35 columns have integer values and 43 are object type

```python
# To check null values 
df.isnull()
```

```python
# To count number of null values any column has
df.isnull().sum()
```

```python
# To remove null values from a columns using fillna 
df['PropertyFrontage'] = df['PropertyFrontage'].fillna(df['PropertyFrontage'].mean())
```

```python
df.isnull().sum()
```

```python
df.shape
```

```python
# Remove rows that have less than a certain number of non-null values
df = df.dropna(thresh=2)
```

```python
df.shape
```

```python
# Remove columns that have NAN values more than 50 percent of the total rows
threshold = 0.5 * len(df)
df = df.loc[:, df.isnull().sum() <= threshold]
```

```python
df.shape
```

# Bivariate Analysis

## Common types of graphs for bivariate analysis :   1. 
Scatter plots: Used for two continuous variables.   2. 
Box plots: Used when one variable is categorical and the other continuous   3. .
Mosaic plots: Used when both variables are categorica   4. l.
Stacked, grouped, or segmented bar charts: Used for the relationship between two categorical variables.s.

```python
plt.scatter(df['PropertySize'], df['PropPrice'])
plt.xlabel('PropertySize')
plt.ylabel('PropPrice')
plt.title('Scatter Plot of ProperrtySize vs PropPrice')
```

## Maximum property lies in the size 50k square feet area and the Property Price data shows that proprety price merely depends upon its size

```python
# Line Plot
plt.plot(df['PropertySize'], label='PropertySize')
plt.plot(df['PropPrice'], label='PropPrice')
plt.xlabel('Index')
plt.ylabel('Values')
plt.title('Line Plot of PropertySize and PropPrice')
plt.legend()
plt.show()
```

```python
plt.scatter(df['YearBuilt'], df['PropPrice'])
plt.xlabel('YearBuilt')
plt.ylabel('PropPrice')
plt.title('Scatter Plot of YearBuilt vs PropPrice')
plt.show()
```
